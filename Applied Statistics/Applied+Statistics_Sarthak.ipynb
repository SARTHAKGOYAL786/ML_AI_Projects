**Question A : Probability that none of the items are defective?**

```python
# As per given data probabilty of a product defective will be 0.05
p   =  0.05
# As given number of samples we take out is 10
n   =  10
# Number of successful trials i.e number of defect will be found is zero
k   =  0
# With help of binomial distribution
from scipy import stats
binomial = stats.binom.pmf(k,n,p)
print('A) All product are good and has no defect %2.4f'%binomial)
```

```python
#Probability of 2 or less defective will be = P(0 defective)+ P(1 defective)+ P(2 defective)
two_defective = stats.binom.pmf(2,10,0.05)
Probability3 = binomial+one_defective+two_defective
print("Probability of 2 or less defective will be %2.3f" %Probability3)
```

**Answer 2C:** Probability that two or fewer of the items are defective is 0.988 or 98.8%(*approx*)

**Statement 3:** A car salesman sells on an average 3 cars per week.

**3.A. What is Probability that in a given week he will sell some cars?**

```python
# As per given data rate = 3 i.e average sales of car per week
rate = 3
# As per possion distribution p= stats.poisson.pmf(n,rate)
# Since We need to find  Probability that in a given week he will sell some cars ,it means atleast one car should be sold out
# i.e. probabilty of atleat one car = 1 - probility(no car sold out)
n=0
possion1 = 1-stats.poisson.pmf(n,rate)
print(" Probability that in a given week he will sell some cars %2.4f"%possion1)
```

**Answer 3A:** Probability that in a given week he will sell some cars is 0.9502 or 95.02%

**3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars?**

```python
# As per question we need to find probabity of 2 or 3 or 4 cars sell out in a week
#  Probability that in a given week he will sell 2 or more but less than 5 cars = Probability of 2 cars sell out + Probability of 3 cars sell out + Probability of 4 cars sell out

# Probability of 2 cars sell out ,n will be 2
A = stats.poisson.pmf(2,rate)
# Probability of 3 cars sell out ,n will be 3
B = stats.poisson.pmf(3,rate)
# Probability of 4 cars sell out ,n will be 4
C = stats.poisson.pmf(4,rate)

possion2 = A+B+C
print("Probability that in a given week he will sell 2 or more but less than 5 cars %2.4f" %possion2)
```

**Answer 3B:** Probability that in a given week he will sell 2 or more but less than 5 cars is 0.6161 or 61.61%

**A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?**

**B. Write the null hypothesis and alternative hypothesis.**

Null Hypothesis (H0): It states that their is no significant difference in funds raised between companies that are still operating and companies that closed down.

Alternative Hypothesis (H1): It state that their is significant difference in funds raised between companies that are still operating and companies that closed down.

**C. Test for significance and conclusion**

```python
# for Test for significance  we need to conclude T-test on difference in funds raised between companies that are still operating and companies that have closed down
# Since we already define the Null Hypothesis and Alternative Hypothesis
# Let perform T- test on alpha = 0.05
import scipy.stats as stats

# Separate the data into two groups
operating = company[company['OperatingState'] == 'Operating']['Funds_in_million']
closed = company[company['OperatingState'] == 'Closed']['Funds_in_million']

# Perform the t-test
t_stat, p_value = stats.ttest_ind(operating, closed, equal_var=False)

# Set the significance level
alpha = 0.05

# Make a conclusion based on the p-value
if p_value < alpha:
    print("There is a significant difference in funds raised between operating and closed companies.")
else:
    print("There is no significant difference in funds raised between operating and closed companies.")
```

So, Conclusion state that there is a significant difference in funds raised between operating and closed companies.

**G. Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:**

Null Hypothesis (H0): The proportion of companies that are operating is the same between winners and contestants.

H0: P1 = P2

Alternative Hypothesis (Ha): The proportion of companies that are operating is different between winners and contestants.

Ha: P1 ≠ P2

Where:
P1 represents the proportion of companies that are winners.
P2 represents the proportion of companies that are contestants.

**H. Test for significance and conclusion**

```python
# for Test for significance  we need to conclude Chi-test on hypothesis comparing the proportion of companies that are operating between winners and contestants
# Since we already define the Null Hypothesis and Alternative Hypothesis
# Let perform Chi- test on alpha = 0.05
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Filter the DataFrame to include only winners and contestants
winners_contestants = company1[company1['Result'].isin(['Winner', 'Contestant'])]

# Create a contingency table of 'Result' vs. 'OperatingState'
contingency_table = pd.crosstab(winners_contestants['Result'], winners_contestants['OperatingState'])

# Perform the chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Set the significance level (alpha)
alpha = 0.05

# Compare the p-value to alpha
if p < alpha:
    print("Reject the null hypothesis: There is a significant difference in proportions.")
else:
    print("Fail to reject the null hypothesis: No significant difference in proportions.")
```
