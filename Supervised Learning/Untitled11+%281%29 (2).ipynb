**4. Model Building: [6 Marks]**

**A. Split data into X and Y. [1 Marks]**

```python
x = comb_file.drop(columns='Class')
y = comb_file["Class"] # target

```

**B. Split data into train and test with 80:20 proportion. [1 Marks]**

```python
#split the data into train and test
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
X_Scaled=x.apply(zscore)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

**C. Train a Supervised Learning Classification base model using KNN classifier. [2 Marks]**

```python
KNN = KNeighborsClassifier(n_neighbors= 5 , metric = 'euclidean' )

KNN.fit(x_train, y_train)
predicted_labels = KNN.predict(x_test)
print('Accuracy on Training data:',KNN.score(x_train, y_train) )
print('Accuracy on Test data:',KNN.score(x_test, y_test) )
```

**D. Print all the possible performance metrics for both train and test data.**

```python
from sklearn.metrics import classification_report
# Predict on training and test data
y_train_pred = KNN.predict(x_train)
y_test_pred = KNN.predict(x_test)

# Generate classification reports for training and test data
train_report = classification_report(y_train, y_train_pred)
test_report = classification_report(y_test, y_test_pred)

# Print the reports
print("Classification Report for Training Data:")
print(train_report)

print("\nClassification Report for Test Data:")
print(test_report)
```
